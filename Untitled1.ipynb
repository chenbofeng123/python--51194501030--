```python
import numpy as np
import pandas as pd
import copy
import time
import warnings
warnings.filterwarnings('ignore')
import gc
gc.collect()
def  drop_series_zero(df):
    drop_cols=[]
    drop_col=[]
    count=0
    
    for idx,row in df.iterrows():
       
        if(row['speed']!=0):
            if(len(drop_col)>50):
                drop_cols.extend(drop_col)
            drop_col=[]
        else:
            drop_col.append(idx)
    #print('删除',len(drop_cols))
    df.drop(drop_cols)
    df=df.reset_index(drop=True)
    return df
def get_hour_seg(v):
    if v<6:
        return 0
    elif v>22:
        return 0
    elif v>=6 and v<11:
        return 1
    elif v>=11 and v<16:
        return 2
    else:
        return 3
def get_week_seg(v):
    if v>=0 and v<=4:
        return 0
    else:
        return 1
#处理经纬度  #计算相邻2个记录之间的距离  变化
from math import radians, atan, tan, sin, acos, cos

def getDistance(latA, lonA, latB, lonB):
    ra = 6378140  # radius of equator: meter  
    rb = 6356755  # radius of polar: meter  
    flatten = (ra - rb) / ra  # Partial rate of the earth  
    # change angle to radians  
    radLatA = radians(latA)
    radLonA = radians(lonA)
    radLatB = radians(latB)
    radLonB = radians(lonB)

    try:
        pA = atan(rb / ra * tan(radLatA))
        pB = atan(rb / ra * tan(radLatB))
        x = acos(sin(pA) * sin(pB) + cos(pA) * cos(pB) * cos(radLonA - radLonB))
        c1 = (sin(x) - x) * (sin(pA) + sin(pB)) ** 2 / cos(x / 2) ** 2
        c2 = (sin(x) + x) * (sin(pA) - sin(pB)) ** 2 / sin(x / 2) ** 2
        dr = flatten / 8 * (c1 - c2)
        distance = ra * (x + dr)
        return distance  # meter   
    except:
        return 0.0000001
def fun(x):
    if(x["pre_latitude"]==0 and x["pre_longitude"]==0):
        return 0
    else: 
        return  getDistance(x["latitude"],x["longitude"],x["pre_latitude"],x["pre_longitude"])    

    
def FE_1(df):
    #处理时间
    df['time']=pd.to_datetime(df['time'])
    df['time_diff']=df['time']-df['time'].shift(1)
    df['time_diff']=df['time_diff'].dt.seconds       
    df['time_diff'].fillna(0,inplace=True)       #和上条记录的时间差
   
    
    df['seg_time_gap']=df['time']-df['time'].iloc[0]  #和首条记录的时间差
    df['seg_time_gap']=df['seg_time_gap'].dt.seconds
    df['month']=df['time'].dt.month
    df['week_day']=df['time'].dt.dayofweek
    
    df['hour']=df['time'].dt.hour
    df['hour_seg']=df['hour'].apply(get_hour_seg)
    df['week_seg']=df['week_day'].apply(get_week_seg)
    df['day']=df['time'].dt.day
    df['minute']=df['time'].dt.minute
    
    #处理速度
    df['speed_diff']=df['speed']-df['speed'].shift(1)  #和上条速度差距
    df['speed_diff'].fillna(0,inplace=True)
    
    df['acce']=df['speed_diff']/df['time_diff']   
    df['acce'].fillna(0,inplace=True)
    
    
    #处理里程
    df['mile_diff']=df['mileage']-df['mileage'].shift(1)  #和上一条的里程差
    df['mile_diff'].fillna(0,inplace=True)
    df['seg_mile_gap']=df['mileage']-df['mileage'].iloc[0]  #和第一条的里程差
    
    
    #处理经纬度
    temp=pd.DataFrame()
    temp['latitude']=df['latitude']
    temp['longitude']=df['longitude']
    temp['pre_latitude']=temp['latitude'].shift(1)
    temp['pre_longitude']=temp['longitude'].shift(1)
    temp.fillna(0,inplace=True)
    df['cal_dis']=temp.apply(fun, axis=1)
    
    
    
    
  
    
    df['soc_gap'] =df['soc']-df['soc'].shift(1) #和上一个soc的差距
    df['soc_gap'].fillna(0,inplace=True)
    
#     df['soc_gap'] =df['soc']-df['soc'].iloc[0] #和第一个soc的差距
#     df['soc_gap'].fillna(0,inplace=True)
    for col in ['speed','speed_diff','time_diff','seg_time_gap','mile_diff','seg_mile_gap']:
            df['{}_distriction_mean'.format(col)] = df[col].rolling(window=10).mean()
            df['{}_distriction_max'.format(col)] =  df[col].rolling(window=10).max() 
            df['{}_distriction_min'.format(col)] =  df[col].rolling(window=10).min() 
            df['{}_distriction_std'.format(col)] =  df[col].rolling(window=10).std() 
            df['{}_distriction_var'.format(col)] =  df[col].rolling(window=10).var()
 
    for col in ['hour','day','minute']:
        for col2 in ['speed','latitude','longitude']:
            df['{}_{}_min'.format(col,col2)] = df.groupby(col)[col2].min()
            df['{}_{}_max'.format(col,col2)]=df.groupby(col)[col2].max()
            df['{}_{}_mean'.format(col,col2)]=df.groupby(col)[col2].mean()
            df['{}_{}_std'.format(col,col2)]=df.groupby(col)[col2].std()

    
    #初始soc
    df['start_soc']=df['soc'].iloc[0]
    df.fillna(0,inplace=True)
    return df
def FE(df):
    df=FE_1(df)
    return df
    
```

```python
import gc
gc.collect()
count =0
alldata=pd.DataFrame()
pianduan=0
def train_process(car_number,count):
    raw_data=pd.read_csv('traindata/Vehicle No.{}.csv'.format(car_number))
    print("--------------处理----{}-------------".format(car_number))

    # =============================================================================
    raw_data = raw_data.dropna(axis=0,how='any')  # 删除包含 NaN 的行
    print('--------------原始数据共有%i条-------------' % raw_data.values.shape[0])
    print('--------------原始数据含有缺失值的%i条-------------' % raw_data.isnull().sum()[0])
   
    # =============================================================================
    # 删除时间重复的值
    # =============================================================================
    a = copy.copy(raw_data)
    raw_data = raw_data.drop_duplicates(subset='time', keep='first', inplace=False)  #删除time相同的索引
    raw_data = raw_data.reset_index(drop=True)
    #print('现在数据共有%i条' % raw_data.values.shape[0])
    print('索引重复数据有%i条' % (len(a) - raw_data.values.shape[0]))

    # =============================================================================
    # 按时间排序
    # =============================================================================
    raw_data = raw_data.sort_values(['time'], ascending = True).reset_index(drop=True)
    print('按时间排序')
    raw_data = raw_data.reset_index().drop('index',axis=1)
    raw_data_copy = copy.deepcopy(raw_data)
    # =============================================================================
    # 计算相邻里程，删除里程差为负的
    # =============================================================================
    raw_data_copy['mileage'] = raw_data_copy['mileage'] - raw_data_copy['mileage'].shift(1)
    clean_data1 = raw_data_copy.fillna(value = 0)
    raw_data = raw_data[clean_data1['mileage'] >= 0]
    raw_data = raw_data.reset_index(drop=True)
    m = raw_data.values.shape[0] - clean_data1.values.shape[0]
    print('里程差为负数的异常数据有%i条' % - m)
    
    
    
    ###  按照  手动充电的那块进行截断分片
    # 速度为0，电流为负，电量增加。 
    # 由于时间粒度非常细，相邻2个电量不一定增加，但是只要 手动充电了，肯定是个时间段，
    # 这个时间块的最大最小时间差肯定是大于几分钟的。


    raw_data['time'] = raw_data['time'].astype(str).apply(lambda x: x[:4]+'-'+x[4:6]+'-'+x[6:8]+' '+x[8:10]+':'+x[10:12]+':'+x[12:])
    raw_data['time'] = pd.to_datetime(raw_data['time']) 
    print('开始实际',raw_data['time'].min())
    print('开始实际',raw_data['time'].max())
    
    #识别出充电时间段
    data= copy.deepcopy(raw_data)
    data = data[data['speed']==0]
    data = data[data['current']<0]
    data['month']=data['time'].dt.month
    data['day']=data['time'].dt.day
    data['hour']=data['time'].dt.hour
    data['minute']=data['time'].dt.minute
    #充电过程是持续的，一定要大于某个值。
    data['shifttime']=data['time']-data['time'].shift(1)
    data['shifttime']=data['shifttime'].apply(lambda x:x.seconds)
    print("可能的充电的片段的长度1:" ,data.shape[0])
    data2=data[data['shifttime']>100] 
    #这些是可能的开始充电点。 这条记录满足充电的条件，并且和上一条处于充电状态，记录相差100s，那么
    data=data.reset_index()
    data['last_index']= data['index'].shift(1)
    data['next_index']= data['index'].shift(-1)
    data2=data[data['shifttime']>100]#last_index 是上一次充电的结束点。 index是这次充电开始的时间点。 那么区间【last_index,index】是运行区间
    print("可能的充电的片段的长度1:" ,data2.shape[0])
    data2=data2.rename(columns={'index':'end','last_index':'start'})
    
    data3=data2[['start','end']]
    data3.loc[0]=[0,data['index'][0]]
    data3=data3.sort_index()
    start = data3['start'].values
    end = data3['end'].values
    #保存了片段
    data = copy.deepcopy(raw_data)
    print("提取出来的充电时间段：",start.shape[0])
    traindata=pd.DataFrame()
    lessthan100=0  #长度小于100
    pianduan          #再次切分片段的时候，增加了的片段。
    for i in range(start.shape[0]):
        first =int(start[i])
        last= end[i]
        tmp = pd.DataFrame(data[first:last])
        
        tmp.fillna(0,inplace=True)
        if(tmp.shape[0]<100):
            lessthan100+=1
            continue
        tmp=tmp[~((tmp['soc']==0)&(tmp['speed']==0)&(tmp['current']==0))]
        tmp = tmp.reset_index()
        tmp =tmp.drop('index',axis=1)
        tmp['shift_time'] = tmp['time']-tmp['time'].shift(1)
        tmp['shift_time'].fillna(0,inplace=True)
        
        s=tmp[tmp['shift_time'].dt.seconds>100].index.to_list()
        
        tmp=tmp.drop(columns='shift_time',axis=1)
       
        if(len(s)==0):
           
            
            tmp['car_number']=car_number
            tmp['pianduan']=count
            traindata=traindata.append(tmp,ignore_index=True)
            count+=1
        else:
            
            last_j=tmp.index.to_list()[0]
            for j in s:
               
                tmp2 = pd.DataFrame(tmp.iloc[last_j:j])
                last_j=j
                tmp2=drop_series_zero(tmp2)

                tmp2['car_number']=car_number
                tmp2['pianduan']=count
                traindata=traindata.append(tmp2,ignore_index=True)
                count+=1
     
     
    print("提取出来的行驶段：",traindata['pianduan'].drop_duplicates().shape[0])
    gc.collect()
    return traindata,count

    
```
